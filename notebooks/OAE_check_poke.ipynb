```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('comparison_data.pkl')
df
```

```python
not_ok = []
count = 0
for n in range(len(df)):
    if df.iloc[n].rmse[-1] > 5:
        info = {k: v for k, v in zip(['polygon', 'basin', 'start_date'], df.iloc[n].name)}
        info["polygon_id"] = df.iloc[n].polygon_id
        info["rmse"] = df.iloc[n].rmse[-1]
        not_ok.append(info)

    plt.plot(df.iloc[n].rmse, label=f'{df.iloc[n].name[1]}-{df.iloc[n].name[2]}')
    count += 1
plt.ylabel('RMSE [mmol m$^{-3}$]')
plt.xlabel('Month since release');
#plt.ylim((0, 100))
count
```

The following simulations have high RMSE:

```python
df_problem = pd.DataFrame(not_ok)
df_problem.to_csv('problem-simulations.csv')
df_problem
```
