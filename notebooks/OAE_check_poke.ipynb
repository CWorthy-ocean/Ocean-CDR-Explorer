```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('comparison_data.pkl')
df
```

```python
not_ok = []
count = 0
for n in range(len(df)):
    if df.iloc[n].rmse[-1] > 50:
        not_ok.append(n)

    plt.plot(df.iloc[n].rmse, label=f'{df.iloc[n].name[1]}-{df.iloc[n].name[2]}')
    count += 1
plt.ylabel('RMSE [mmol m$^{-3}$]')
plt.xlabel('month');
count
```

The following simulations have high RMSE:

```python
len(not_ok)
```

```python
for n in not_ok:
    print(f'{df.iloc[n].name}: {df.iloc[n].rmse[-1]:0.2f}')
```
