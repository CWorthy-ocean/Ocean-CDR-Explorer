```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('comparison_data.pkl')
df
```

```python
for n in range(len(df)):
    if df.iloc[n].rmse[-1] == 0.0: #0.5e8:
        print(df.iloc[n].name)
    plt.plot(df.iloc[n].rmse, label=f'{df.iloc[n].name[1]}-{df.iloc[n].name[2]}')

plt.ylabel('RMSE')
plt.xlabel('month');
```
